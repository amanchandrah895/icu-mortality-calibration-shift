```python
# =========================
# Cell 1
# Load dataset
# =========================

import pandas as pd

path = "/content/drive/MyDrive/Plan A/final/mimiciv_final_dataset_clean.csv"
df = pd.read_csv(path)

print("Shape:", df.shape)
df.head()
```

```python
# =========================
# Cell 2
# Separate columns properly
# =========================

id_cols = ["subject_id", "hadm_id", "stay_id"]
time_cols = ["intime", "outtime"]
target_col = ["mortality"]

# Identify numeric columns ONLY
numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()

# Remove target from numeric features
numeric_cols = [col for col in numeric_cols if col not in target_col]

# Categorical columns
categorical_cols = ["gender"]

print("Numeric features:", len(numeric_cols))
print("Categorical features:", categorical_cols)
```

```python
# =========================
# Cell 3 (UPDATED)
# Missing indicators
# =========================

for col in numeric_cols:
    df[col + "_missing"] = df[col].isnull().astype(int)
```

```python
# =========================
# Cell 4 (UPDATED)
# Median imputation
# =========================

for col in numeric_cols:
    median = df[col].median()
    df[col] = df[col].fillna(median)
```

```python
# =========================
# Encode gender FIRST
# =========================

df["gender"] = df["gender"].map({"M": 1, "F": 0})
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
# =========================
# Final NaN cleanup
# =========================

# Fill any remaining NaNs safely
df = df.fillna(0)

print("Remaining missing:", df.isnull().sum().sum())
```

```python
# =========================
# FINAL SAVE (MANDATORY)
# =========================

save_path = "/content/drive/MyDrive/Plan A/final/mimiciv_processed.csv"
df.to_csv(save_path, index=False)

print("Saved processed dataset:", save_path)
```
